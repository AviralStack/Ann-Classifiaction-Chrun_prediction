```python
import torch
import torch.nn as nn

class AnnModel(nn.Module):
    def __init__(self, input_dim=12): 
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, 8),
            nn.ReLU(),
            nn.Linear(8, 4),
            nn.ReLU(),
            nn.Linear(4, 1),
            nn.Sigmoid()
        )
    def forward(self, features):
        return self.network(features)
```

```python

model = AnnModel(input_dim=12)


model.load_state_dict(torch.load('ann_model_weights.pth'))

model.eval()
```

```python
import numpy as np

sample_data = np.array([0.5, 1.2, 0.1, 0.9, 0.4, 0.2, 0.7, 0.3, 0.6, 1.1, 0.0, 0.5])


with torch.no_grad():
    input_tensor = torch.from_numpy(sample_data).float().unsqueeze(0) # add batch dim
    output = model(input_tensor)
    
    
    prob = output.item()
    prediction = 1 if prob >= 0.5 else 0

print(prediction)
```
